```python
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings("ignore")

```

```python
df = pd.read_csv('Adidas Sales.csv')
```

```python
df.head()
```

```python
# Cleaning Date Column
df['Invoice Date'] = pd.to_datetime(df['Invoice Date']).dt.date
```

```python
df.head()
```

```python
# Cleaning [Price Per Unit Column] -- Removing($)
df['Price per Unit'] = df['Price per Unit'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

```python
# Cleaning [Total Sales] -- Removing (,)
df['Total Sales'] = df['Total Sales'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

```python
df['Operating Profit'] = df['Operating Profit'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

```python
# Cleaning [Units Sold] and converting it into Integer
df["Units Sold"]=df["Units Sold"].replace(',','',regex=True).astype(int)
```

```python
df.head()
```

```python
# Saving Cleaned File
df.to_csv("Adidas_Sales_Cleaned.csv", index=False)
```
